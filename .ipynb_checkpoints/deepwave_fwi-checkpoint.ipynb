```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
import deepwave as dw
```

```python
device = 'cuda'
```

```python
nx = 128
ny = 32
dx = 5.

vp_background = torch.ones(ny, nx, device=device) * 1500
vs_background = torch.ones(ny, nx, device=device) * 1000
rho_background = torch.ones(ny, nx, device=device) * 2200
```

```python
vp_true = vp_background.clone()
vp_true[10:20, 30:40] = 1600
vs_true = vs_background.clone()
vs_true[10:20, 45:55] = 1100
rho_true = rho_background.clone()
rho_true[10:20, 60:70] = 2300
```

```python
fig, ax = plt.subplots(1, 3, figsize=(20, 5))
ax[0].imshow(vp_true.cpu(), aspect='auto')
ax[1].imshow(vs_true.cpu(), aspect='auto')
ax[2].imshow(rho_true.cpu(), aspect='auto')
plt.show()
```

```python
n_shots = 8
n_sources_per_shot = 1
d_source = 12 
first_source = 8
source_depth = 1

n_receivers_per_shot = nx-1
d_receiver = 1
first_receiver = 0
receiver_depth = 1
```

```python
freq = 15
nt = 200
dt = 0.004
peak_time = 1.5 / freq
```

```python
source_locations = torch.zeros(n_shots, n_sources_per_shot, 2, dtype=torch.float32, device=device)
source_locations[..., 0] = source_depth
source_locations[:, 0, 1] = (torch.arange(n_shots) * d_source + first_source)
```

```python
receiver_locations = torch.zeros(n_shots, n_receivers_per_shot, 2, dtype=torch.float32, device=device)
receiver_locations[..., 0] = receiver_depth
receiver_locations[:, :, 1] = (torch.arange(n_receivers_per_shot) * d_receiver + first_receiver).repeat(n_shots, 1)
```

```python
source_amplitudes = (dw.wavelets.ricker(freq, nt, dt, peak_time)).repeat(n_shots, n_sources_per_shot, 1).to(device)
```

```python
observed_data = torch.stack(dw.elastic(*dw.common.vpvsrho_to_lambmubuoyancy(vp_true, vs_true, rho_true), dx, dt,
                        source_amplitudes_y=source_amplitudes,
                        source_locations_y=source_locations,
                        receiver_locations_x=receiver_locations,
                        receiver_locations_y=receiver_locations,
                        pml_width=[0, 20, 20, 20]
)[-2:])
```

```python
observed_data.shape
```

```python
vp = vp_background.clone().requires_grad_()
vs = vs_background.clone().requires_grad_()
rho = rho_background.clone().requires_grad_()
optimiser = optim.LBFGS([vp, vs, rho])
loss_fn = nn.MSELoss()

n_epochs = 20

for epoch in tqdm(range(n_epochs)):
    def closure():
        optimiser.zero_grad()
        out = torch.stack(dw.elastic(
            *dw.common.vpvsrho_to_lambmubuoyancy(vp, vs, rho),
            dx, dt,
            source_amplitudes_y=source_amplitudes,
            source_locations_y=source_locations,
            receiver_locations_x=receiver_locations,
            receiver_locations_y=receiver_locations,
            pml_width=[0, 20, 20, 20]
        )[-2:])
        loss = 1e20*loss_fn(out, observed_data)
        loss.backward()
        return loss

    optimiser.step(closure)
```

```python
vp_true.shape
```

```python
import torch.nn.functional as F

class VelocityNet(nn.Module):
    def __init__(self, n_shots, nx, ny, nt):
        super().__init__()
        self.c1 = nn.Conv2d(in_channels=n_shots*2, out_channels=1, kernel_size=3, padding=1)
        self.l1 = nn.Linear(nt, nx)
        self.l2 = nn.Linear(nx-1, nx)
        
    def forward(self, x):
        x = F.relu(self.c1(x.reshape(n_shots*2, nx-1, nt)).squeeze())
        x = F.relu(self.l1(x)).permute(1, 0)
        x = F.relu(self.l2(x))
        return x 

class VelocityModel(nn.Module):
    def __init__(self, n_shots, nx, ny, nt):
        super().__init__()
        self.vp_net = VelocityNet(n_shots, nx, ny, nt)
        self.vs_net = VelocityNet(n_shots, nx, ny, nt)
        self.rho_net = VelocityNet(n_shots, nx, ny, nt)
        
    def forward(self, x):
        vp = self.vp_net(x)
        vs = self.vs_net(x)
        rho = self.rho_net(x)
        return vp, vs, rho

model = VelocityModel(n_shots, nx, ny, nt).to(device)
loss_fn = nn.MSELoss()
```

```python
optimizer = optim.AdamW(model.parameters(), lr=.1)
lr_scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=10, factor=0.75)
n_epochs = 1000

for epoch in (pbar:=tqdm(range(n_epochs))):
    optimiser.zero_grad()
    
    vp, vs, rho = model(observed_data)
    out = torch.stack(dw.elastic(
        *dw.common.vpvsrho_to_lambmubuoyancy(vp, vs, rho),
        dx, dt,
        source_amplitudes_y=source_amplitudes,
        source_locations_y=source_locations,
        receiver_locations_x=receiver_locations,
        receiver_locations_y=receiver_locations,
        pml_width=[0, 20, 20, 20]
    )[-2:])
    
    loss = 1e10*loss_fn(out, observed_data)

    loss.backward()
    optimizer.step()
    lr_scheduler.step(loss)

    pbar.set_description(f"Epoch {epoch}, Loss: {loss.item():.6e}")
```

```python
vp_hat, vs_hat, rho_hat = model(observed_data)
fig, ax = plt.subplots(1, 3, figsize=(20, 5))
ax[0].imshow(vp_hat.detach().cpu(), aspect='auto')
ax[1].imshow(vs_hat.detach().cpu(), aspect='auto')
ax[2].imshow(rho_hat.detach().cpu(), aspect='auto')
plt.show()
```
