```python
from PS_Separated_utils import *
import torch.nn as nn
import torch.optim as optim
torch.set_default_device("cpu")
```

```python
nx = 256
ny = 64
dx = 0.03
dt = 0.005
nt = 200
damping_size = 32
alpha = add_random_shapes(torch.empty((ny, nx)), base_value=4., num_shapes=20, damping_size=damping_size)
beta = add_random_shapes(torch.empty((ny, nx)), base_value=2., num_shapes=20, damping_size=damping_size)
alpha[-ny//10:] = 0
beta[-ny//10:] = 0
beta[beta>=alpha] /= 2.
```

```python
#v_xp, v_xs, v_zp, v_zs, u, w = initialise_conditions(nx, ny, (nx//2, 0))
#conditions = (v_xp, v_xs, v_zp, v_zs, u, w)
#trajectories = simulate(v_xp, v_xs, v_zp, v_zs, u, w, alpha, beta, dx, nx, ny, dt, nt, damping_size).unsqueeze(0)

positions = [(nx//3, 0), (3*nx//3, 0)]
trajectories = []
conditions = []
for pos in positions:
    v_xp, v_xs, v_zp, v_zs, u, w = initialise_conditions(nx, ny, pos)
    conditions.append((v_xp, v_xs, v_zp, v_zs, u, w))
    trajectories.append(simulate(v_xp, v_xs, v_zp, v_zs, u, w, alpha, beta, dx, nx, ny, dt, nt, damping_size))
trajectories = torch.stack(trajectories)
```

```python
import torch
import torch.nn as nn

class Dynamics(nn.Module):
    def __init__(self, nx, ny, nt, damping_size):
        super().__init__()
        self.sigma = damping(nx, ny, damping_size)
        self.alpha_net = nn.Sequential(
            nn.Conv2d(in_channels=2, out_channels=16, kernel_size=3, padding=1),
            nn.Conv2d(in_channels=16, out_channels=ny, kernel_size=3, padding=1)
        )
        self.alpha_net_2 = nn.Sequential(
            nn.Linear((nt)//10 +1, 1),
            nn.ReLU()
        )
        
        self.beta_net = nn.Sequential(
            nn.Conv2d(in_channels=2, out_channels=16, kernel_size=3, padding=1),
            nn.Conv2d(in_channels=16, out_channels=ny, kernel_size=3, padding=1)
        )
        self.beta_net_2 = nn.Sequential(
            nn.Linear((nt)//10 +1, 1),
            nn.ReLU()
        )

    def get_alpha(self, trace):
        x = trace.permute(2, 1, 0).unsqueeze(0)
        return self.alpha_net_2(self.alpha_net(x)).squeeze()

    def get_beta(self, trace):
        x = trace.permute(2, 1, 0).unsqueeze(0)
        return self.beta_net_2(self.beta_net(x)).squeeze()
        
    def forward(self, trace, v_xp, v_xs, v_zp, v_zs, u, w, dx, dt):
        x = trace.permute(2, 1, 0).unsqueeze(0)
        alpha = self.alpha_net_2(self.alpha_net(x)).squeeze()
        beta = self.beta_net_2(self.beta_net(x)).squeeze()
        du_dx, du_dz = gradient(u, dim=(0, 1), spacing=dx)
        dw_dx, dw_dz = gradient(w, dim=(0, 1), spacing=dx)

        A = du_dx.add(dw_dz)
        B = du_dz.sub(dw_dx)

        dA_dx, dA_dz = gradient(A, dim=(0, 1), spacing=dx)
        dB_dx, dB_dz = gradient(B, dim=(0, 1), spacing=dx)
        alpha_sq = torch.square(alpha)
        beta_sq = torch.square(beta)
        v_xp.addcmul_(alpha, dA_dx, value=dt)
        v_xs.addcmul_(beta_sq, dB_dz, value=dt)
        v_zp.addcmul_(alpha_sq, dA_dz, value=dt)
        v_zs.addcmul_(beta_sq, dB_dx, value=-dt)
        
        u.add_(v_xp + v_xs, alpha=dt)
        w.add_(v_zp + v_zs, alpha=dt)

        damping_factor = 1 - self.sigma
        v_xp.mul_(damping_factor)
        v_xs.mul_(damping_factor)
        v_zp.mul_(damping_factor)
        v_zs.mul_(damping_factor)

        return v_xp, v_xs, v_zp, v_zs, u, w
```

```python
def simulate_optimisation(dynamics, trace, v_xp, v_xs, v_zp, v_zs, u, w, dx, dt, nt, pbar, epoch):
    trajectory = [torch.concatenate((u.unsqueeze(-1), w.unsqueeze(-1)), axis=-1).unsqueeze(0)]
    for it in (range(nt)):
        pbar.set_postfix_str(f"{it}/{nt}, {epoch}")
        v_xp, v_xs, v_zp, v_zs, u, w = dynamics(trace, v_xp, v_xs, v_zp, v_zs, u, w, dx, dt)
        trajectory.append(torch.concatenate((u.unsqueeze(-1), w.unsqueeze(-1)), axis=-1).unsqueeze(0))
    return torch.vstack(trajectory)
```

```python
torch.manual_seed(0)
dynamics = Dynamics(nx, ny, nt, damping_size)
optimizer = optim.AdamW(dynamics.parameters(), lr=0.002)

loss_fn = nn.MSELoss()
losses = []
epochs = 1000

for epoch in (pbar:=tqdm(range(epochs))):
    optimizer.zero_grad()
    l = 0
    for i, c in enumerate(conditions):
        v_xp, v_xs, v_zp, v_zs, u, w = c
        u_sim = simulate_optimisation(dynamics, trajectories[i, ::10, 0], v_xp.clone(), v_xs.clone(), v_zp.clone(), v_zs.clone(), u.clone(), w.clone(), dx, dt, nt, pbar, epoch)
        loss = loss_fn(trajectories[i, :, 0], u_sim[:, 0]).clamp(0, 1e2)
        l += loss
    l.backward()
    losses.append(l.item())
    optimizer.step()
    pbar.set_description(f"Epoch {epoch} - nt: {nt} - Loss: {l.item():.5e}")
```
